# exoplanet_schematic

This short tutorial will show you how to make a schematic diagram of an exoplanet system, which can be useful for papers that may benefit from a visual aid.

We first import our necessary modules, including some constants from the `astropy` module:

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.constants import R_sun, au, R_earth
```

The next thing we will add in is our known radius of our host star - in this case, our star in question is an **0.56 RSun**
star, but we need to do some unit conversions for later. We need to convert our radius first into km, and then into AU and Earth radii for our scaling later on.

```python
rs = 0.56 * R_sun.value
print("Stellar radius:", rs, "km")

rs_au = rs / au.value
print("Stellar radius:", rs_au, "AU")

rs_earth = 0.56251 * R_earth.value
print("Stellar radius:", rs_earth, "REarth")
```

Now we need to know the semi-major axes of our planets in AU. In this case, we have a 2-planet system and we have already found our values from modelling.

```python
a_b = 0.04
a_c = 0.07
```

Next we need to set up some linspace coordinates for drawing the planet orbits as circles. These start and end at our negative and positive values of the semi-major axes.

```python
x_b = np.linspace(-a_b, a_b, 1001)
x_c = np.linspace(-a_c, a_c, 1001)
```

Now we can do the same with the y-coordinates:

```python
y_b1 = np.sqrt(a_b**2 - x_b**2)       #  y values for circular orbits
y_b2 = -1 * np.sqrt(a_b**2 - x_b**2)

y_c1 = np.sqrt(a_c**2 - x_c**2)       #  y values for circular orbits
y_c2 = -1 * np.sqrt(a_c**2 - x_c**2)
```

Now we can start plotting! We need 1 plot axis with a figure size of 6 by 6 inches. We'll build this code blocks as we go.

```python
fig, ax = plt.subplots(1, 1, figsize=(6,6))
```

Now we'll plot the circles. What happens if we comment out some of the code lines? How much of the circle will be plotted?

```python
fig, ax = plt.subplots(1, 1, figsize=(6,6))

ax.plot(x_b, y_b1, '#fca80e')
ax.plot(x_b, y_b2, '#fca80e')
ax.plot(x_c, y_c1, '#6998eb')
ax.plot(x_c, y_c2, '#6998eb')
```

Let's add in a marker for our star in the centre. Just an x will do for now.

```python
fig, ax = plt.subplots(1, 1, figsize=(6,6))

ax.plot(x_b, y_b1, '#fca80e')
ax.plot(x_b, y_b2, '#fca80e')
ax.plot(x_c, y_c1, '#6998eb')
ax.plot(x_c, y_c2, '#6998eb')

ax.plot(0, 0, 'kx')
```

Looks pretty good, but now we need some planets! We're going to scale the size of the planet markers relative to each other, so we'll give planet b a scale of 1. We know that the radius of b is **1.4 REarth** and the radius of c is **2.1 REarth**.

We also have to change the position of one of the planets so they don't overlap, and we can adjust the value of `theta_b` or `theta_c` as we like. Have a go at changing the values of these and seeing where the planets move!

Finally, we'll add the planet markers as circles.

```python
fig, ax = plt.subplots(1, 1, figsize=(6,6))

ax.plot(x_b, y_b1, '#fca80e')
ax.plot(x_b, y_b2, '#fca80e')
ax.plot(x_c, y_c1, '#6998eb')
ax.plot(x_c, y_c2, '#6998eb')

ax.plot(0, 0, 'kx')

rp_b = 1
rp_c = 2.1 / 1.4
rs_scaled = rs_earth / 1.39763

theta_b = 330
theta_c = 0
xb, yb = -a_b * np.cos(theta_b / 180 * np.pi), a_b * np.sin(theta_b / 180 * np.pi)
xc, yc = 0, -1*a_c

plt.plot(xb, yb, '#fca80e', marker='o', ms=rp_b*8)  # 'ms' is marker size. Uses the scalled planet radii
plt.plot(xc, yc, '#6998eb', marker='o', ms=rp_c*8)
```

Okay, looks great - but it's probably a good idea to have some annotations! We'll offset them a bit, try changing the values of the coordinates to see where the text moves.

```python
fig, ax = plt.subplots(1, 1, figsize=(6,6))

ax.plot(x_b, y_b1, '#fca80e')
ax.plot(x_b, y_b2, '#fca80e')
ax.plot(x_c, y_c1, '#6998eb')
ax.plot(x_c, y_c2, '#6998eb')

ax.plot(0, 0, 'kx')

rp_b = 1
rp_c = 2.1 / 1.4
rs_scaled = rs_earth / 1.39763

theta_b = 330
theta_c = 0
xb, yb = -a_b * np.cos(theta_b / 180 * np.pi), a_b * np.sin(theta_b / 180 * np.pi)
xc, yc = 0, -1*a_c

plt.plot(xb, yb, '#fca80e', marker='o', ms=rp_b*8)
plt.plot(xc, yc, '#6998eb', marker='o', ms=rp_c*8)

plt.annotate('b', xy=(xb-0.012, yb+0.001),
             fontweight='bold',
             fontsize=26,
             color='#fca80e')
plt.annotate('c', xy=(xc-0.012, yc+0.005),
             fontweight='bold',
             fontsize=26,
             color='#6998eb')
plt.annotate('TOI-836', xy=(xc-0.018, yc+0.08),
            fontweight='bold',
            fontsize=18,
            color='#000000')
```

Finally, let's get rid of the axes and ticks, and add a scale bar.

```python
fig, ax = plt.subplots(1, 1, figsize=(6,6))

ax.plot(x_b, y_b1, '#fca80e')
ax.plot(x_b, y_b2, '#fca80e')
ax.plot(x_c, y_c1, '#6998eb')
ax.plot(x_c, y_c2, '#6998eb')

ax.plot(0, 0, 'kx')

rp_b = 1
rp_c = 2.1 / 1.4
rs_scaled = rs_earth / 1.39763

theta_b = 330
theta_c = 0
xb, yb = -a_b * np.cos(theta_b / 180 * np.pi), a_b * np.sin(theta_b / 180 * np.pi)
xc, yc = 0, -1*a_c

plt.plot(xb, yb, '#fca80e', marker='o', ms=rp_b*8)
plt.plot(xc, yc, '#6998eb', marker='o', ms=rp_c*8)

plt.annotate('b', xy=(xb-0.012, yb+0.001),
             fontweight='bold',
             fontsize=26,
             color='#fca80e')
plt.annotate('c', xy=(xc-0.012, yc+0.005),
             fontweight='bold',
             fontsize=26,
             color='#6998eb')
plt.annotate('TOI-836', xy=(xc-0.018, yc+0.08),
            fontweight='bold',
            fontsize=18,
            color='#000000')

ax.spines['right'].set_color('white')
ax.spines['left'].set_color('white')
ax.spines['top'].set_color('white')
ax.spines['bottom'].set_color('white')
plt.xticks([])
plt.yticks([])

ax.plot([-0.1, -0.05], [-0.11, -0.11], 'k-', lw=4) # This is for the 0.05 AU scale bar
plt.annotate('0.05 AU', xy=(-0.093, -0.105), fontsize=18, fontweight='bold')

plt.subplots_adjust(top=0.98, right=0.98, left=0.02, bottom=0.02, hspace=0., wspace=0.)
```
